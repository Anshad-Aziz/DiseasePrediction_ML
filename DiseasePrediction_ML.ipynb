```python
import xgboost as xgb
from sklearn import linear_model,model_selection
from sklearn.preprocessing import LabelEncoder
```

```python
traindataname=trainingtarget
testdataname=trainingtarget

prognosislabel=LabelEncoder()

trainingtarget=prognosislabel.fit_transform(trainingtarget)
testtarget=prognosislabel.fit_transform(testtarget)
```

```python
model=linear_model.LogisticRegression()
model.fit(trainingdata,trainingtarget)
```

```python
print(model.score(trainingdata,trainingtarget))
```

```python
print(model.score(testdata,testtarget))
```

```python
print(model.score(trainingdata,trainingtarget),model.score(testdata,testtarget))
```

```python
print("The prediction code of first training row:",model.predict([trainingdata.iloc[0]]))
```

```python
print('Decoding the code:',prognosislabel.inverse_transform(model.predict([trainingdata.iloc[0]])))
```

```python
print('the truth of first row:',traindataname.iloc[0])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_predicted=model.predict(testdata)
```

```python
y_predicted=y_predicted.astype(int)
```

```python
testingtarget=testtarget.astype(int)
```

```python
CM=confusion_matrix(testingtarget,y_predicted)
```
