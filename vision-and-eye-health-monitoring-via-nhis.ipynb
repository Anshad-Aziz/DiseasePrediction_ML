```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
df=pd.read_csv('/kaggle/input/nhis-vision-and-eye-health-surveillance-dataset/NHIS_Vision_and_Eye_Health_Surveillance_20240501.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.set_palette(sns.color_palette(["#8B4513", "#A0522D", "#CD853F", "#D2691E", "#8B7E66", "#A67B5B", "#8B5A2B"]))
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='Gender')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Gender',y='Data_Value')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df,x='Age',y='Data_Value')
plt.title('Data Value by Age')
plt.xlabel('Age Group ')
plt.ylabel('Data Value')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Sample_Size',y='Data_Value')
plt.title('Sample size VS Data Value')
plt.xlabel('Sample Size')
plt.ylabel('Data Value')
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['RaceEthnicity'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=sns.color_palette())
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df,x='Age',y='Data_Value',hue='Gender')
plt.title('Data Value by Age and Gender')
plt.xlabel('Age Group')
plt.ylabel('Data Value')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Data_Value',bins=20)
plt.title('Distribution of Data value')
plt.xlabel('Data value')
plt.ylabel('Frequency')
plt.show()
```
