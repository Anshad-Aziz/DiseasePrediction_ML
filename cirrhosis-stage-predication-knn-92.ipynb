```python
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_columns=ds.select_dtypes(include=['object']).columns.tolist()
```

```python
encoder=OneHotEncoder(sparse_output=False)
```

```python
one_hot_encoded=encoder.fit_transform(ds[categorical_columns])
one_hot_df=pd.DataFrame(one_hot_encoded,columns=encoder.get_feature_names_out(categorical_columns))
```

```python
ds_encoded=pd.concat([ds,one_hot_df],axis=1)
ds_encoded=ds_encoded.drop(categorical_columns,axis=1)
```

```python
print(ds_encoded.head())
```

```python
colx=ds_encoded.columns.tolist()
colx.remove('Stage')
```

```python
x = ds_encoded[colx].values
y=ds_encoded["Stage"].values

del colx
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
x_test=sc_X.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
knn=KNeighborsClassifier(n_neighbors=3,metric='minkowski',p=1)
knn.fit(X_train,Y_train)
y_pred=knn.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print('KNN model accuracy:',accuracy)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(classification_report(y_pred,y_test))
```
